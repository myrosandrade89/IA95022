## **Aggregation of firm data per quarter (2000Q1 - 2022Q2)**

```python
# Storing the US firms quarters
us_q = pd.read_csv("dataset/us2022q2a.csv")

# Storing the US firms data
us_firms = pd.read_csv("dataset/usfirms2022.csv")
us_firms = us_firms[["firm", "Name","Sector NAICS\nlevel 1"]]
us_firms["industry"] = us_firms["Sector NAICS\nlevel 1"]
us_firms = us_firms.drop(columns = ["Sector NAICS\nlevel 1"])
```

```python
# Merging the us_q and us_firms DataFrames
us_q = pd.merge(us_q, us_firms, on = ["firm"]) 
us_q
```

---
## **Aggregation of firm data per year (2000 - 2021)**

```python
us_q_agg_sum = us_q.groupby([us_q.firm, us_q.year]).sum().reset_index()
us_q_agg_last = us_q.groupby([us_q.firm, us_q.year]).last().reset_index()

us_y = pd.DataFrame()

us_y["firm"] = us_q_agg_last["firm"]
us_y["name"] = us_q_agg_last["Name"]
us_y["industry"] = us_q_agg_last["industry"]

us_y["year"] = us_q_agg_last["q"].map(lambda x: int(str(x)[:-2]))
us_y["revenue"] = us_q_agg_sum["revenue"]
us_y["cogs"] = us_q_agg_sum["cogs"]
us_y["sgae"] = us_q_agg_sum["sgae"]
us_y["otheropexp"] = us_q_agg_sum["otheropexp"]
us_y["extraincome"] = us_q_agg_sum["extraincome"]
us_y["finexp"] = us_q_agg_sum["finexp"]
us_y["incometax"] = us_q_agg_sum["incometax"]

us_y["totalassets"] = us_q_agg_last["totalassets"]
us_y["totalliabilities"] = us_q_agg_last["totalliabilities"]
us_y["shortdebt"] = us_q_agg_last["shortdebt"]
us_y["longdebt"] = us_q_agg_last["longdebt"]
us_y["stockholderequity"] = us_q_agg_last["stockholderequity"]
us_y["adjprice"] = us_q_agg_last["adjprice"]
us_y["originalprice"] = us_q_agg_last["originalprice"]
us_y["sharesoutstanding"] = us_q_agg_last["sharesoutstanding"]
us_y["sharesoutstanding"] = us_q_agg_last["sharesoutstanding"]

# Calculating the firm size
us_y["market value"] = us_y["originalprice"] * us_y["sharesoutstanding"]
us_y["book value"] = us_y["stockholderequity"]

# Calculating the sales
us_y["sales"] = us_y["revenue"]

# Profit margin measures
us_y["net income"] = us_y["revenue"] + us_y["extraincome"] - us_y["cogs"] - us_y["sgae"] - us_y["otheropexp"] - us_y["finexp"] - us_y["incometax"]
us_y["profit margin"] = us_y["net income"] / us_y["sales"]

# Calculating the yearly cc returns, c proft margins and c book value / market value
us_y_cc = us_y.groupby([us_y.firm], as_index = False)[["adjprice", "profit margin", "book value", "market value", "year", "firm"]].shift(1)
us_y_cc["year"] += 1
us_y = pd.merge(us_y, us_y_cc, on = ["firm", "year"], how = "left")

us_y["r"] = np.log(us_y["adjprice_x"]) - np.log(us_y["adjprice_y"])
us_y["adjprice"] = us_y["adjprice_x"]

us_y["c profit margin"] = us_y["profit margin_x"] - us_y["profit margin_y"]
us_y["profit margin"] = us_y["profit margin_x"]

us_y["cc book/market"] = np.log(us_y["book value_x"] / us_y["market value_x"]) - np.log(us_y["book value_y"] / us_y["market value_y"])
us_y["book value"] = us_y["book value_x"]
us_y["market value"] = us_y["market value_x"]


us_y = us_y.drop(columns = ["adjprice_x", "adjprice_y", "profit margin_x","profit margin_y", "market value_x","market value_y", "book value_x", "book value_y"])

us_y = us_y[us_y.year != 2022]

us_y
```

- How can you measure firm profitability that can be used to compare performance among firms of different sizes? Select and justify at least 3 measures and show descriptive statistics

```python
big_company = us_y[us_y.firm == "MSFT"]
medium_company = us_y[us_y.firm == "XYL"]
small_company = us_y[us_y.firm == "PBT"]
```

FIRST MEASURE: CC RETURNS

IT MEASURES THE CONTINUOUSLY COMPOUNDED CHANGE BETWEEN THE ADJUSTED PRICES EVERY YEAR, THUS IT SHOWS HOW THE RETURNS INCREASE OR DECREASE OVER A PERIOD OF TIME. IT IS GOOD TO MEASURE PERFORMANCE AMONG FIRMS OF DIFFERENCE SIZES SINCE THE CHANGE IS MEASURED ON PERCENTUAL POINTS INDEPENDENT OF ANY RELATION TO THE SIZE.

```python
# Calculating the cc returns
plt.clf()
plt.title("CC Returns")
plt.plot( big_company["year"], big_company["r"], label = "big company")
plt.plot( medium_company["year"], medium_company["r"], label = "medium company")
plt.plot( small_company["year"], small_company["r"], label = "small company")

plt.legend()
plt.show()
```

SECOND MEASURE: C PROFIT MARGIN

IT MEASURES THE CHANGE BETWEEN PROFITS MARGIN EVERY YEARS. SINCE THE PROFIT MARGIN MEASURES THE NET INCOME DIVIDED BY THE SALES, IT IS GOING TO REPRESENT THE RETURN PER DOLLAR GAINED IRRESPECTIVE OF THE FIRM SIZE.

```python
# Calculating the c profit margin
plt.clf()
plt.title("C Profit Margin")
plt.plot( big_company["year"], big_company["c profit margin"], label = "big company")
plt.plot( medium_company["year"], medium_company["c profit margin"], label = "medium company")
plt.plot( small_company["year"], small_company["c profit margin"], label = "small company")

plt.legend()
plt.show()
```

THIRD MEASURE: CC BOOK VALUE / MARKET VALUE

IT MEASURES THE CONTINUOUSLY COMPOUNDED CHANGE OF THE RELATIONSHIP BETWEEN THE BOOK AND MARKET VALUE EVERY YEAR. THE RELATIONSHIP EXPRESSES WHETHER THE ACTUAL PERFORMANCE OF A COMPANY'S VALUE PERFORMS ABOVE OR BELOW MARKET EXPECTATIONS IRRESPECTIVE OF THE FIRM SIZE.

```python
# Calculating the cc book value / market value
plt.clf()
plt.title("CC book/market")
plt.plot( big_company["year"], big_company["cc book/market"], label = "big company")
plt.plot( medium_company["year"], medium_company["cc book/market"], label = "medium company")
plt.plot( small_company["year"], small_company["cc book/market"], label = "small company")

plt.legend()
plt.show()
```

```python
# Descriptive statistics
print("Big company")
print(big_company[["r", "c profit margin", "cc book/market"]].describe())
print("Medium company")
print(medium_company[["r", "c profit margin", "cc book/market"]].describe())
print("Small company")
print(small_company[["r", "c profit margin", "cc book/market"]].describe())
```

WE CAN SEE THAT OUR THREE SELECTED MEASURES ADEQUATELY EXPRESS THE NATURE OF PROFITABILITY ON THE THREE SAMPLE FIRMS SINCE (ALONG WITH THE PLOTS PREVIOUSLY SHOWED) THEY GIVE US A GOOD UNDERSTANDING ON THE MARKET TREND AND SENTIMENT OF THE AFOREMENTIONED.

### 2.2 About statistical modeling

## **History of firm data**

```python
# Calculating variables for the entire history data
us_h = us_q.copy()

# Calculating the size of the firms
us_h["market value"] = us_h["originalprice"] * us_h["sharesoutstanding"]
us_h["book value"] = us_h["stockholderequity"]

# Profit margin measures
us_h["ebit"] = us_h["revenue"] - us_h["cogs"] - us_h["sgae"] - us_h["otheropexp"] 
us_h["net income"] = us_h["ebit"] + us_h["extraincome"] - us_h["finexp"] - us_h["incometax"]
us_h["profit margin"] = (us_h["net income"] / us_h["revenue"]).replace(np.inf, np.nan).replace(-np.inf, np.nan)


# Calculating the EPSP
us_h["eps"] = us_h["net income"] / us_h["sharesoutstanding"]
us_h["epsp"] = us_h["eps"] / us_h["originalprice"]
 
# Calculating the yearly cc returns, c proft margins and c book value / market value
""
us_h_cc = us_h.groupby([us_h.firm], as_index = False)[["adjprice", "profit margin", "book value", "market value"]].shift(4)
us_h = pd.merge(us_h, us_h_cc, left_index=True, right_index=True, how = "left")

# Calculating annual cc returns and stock returns
us_h["r"] = np.log(us_h["adjprice_x"]) - np.log(us_h["adjprice_y"])
us_h["annual stock return"] = (us_h["adjprice_x"] - us_h["adjprice_y"]) / us_h["adjprice_y"]
us_h["adjprice"] = us_h["adjprice_x"]

us_h["c profit margin"] = us_h["profit margin_x"] - us_h["profit margin_y"]
us_h["profit margin"] = us_h["profit margin_x"]

us_h["cc book/market"] = np.log(us_h["book value_x"] / us_h["market value_x"]) - np.log(us_h["book value_y"] / us_h["market value_y"])
us_h["book value"] = us_h["book value_x"]
us_h["market value"] = us_h["market value_x"]

# Calculating the book/market ratio
us_h["book/market"] = np.log(us_h["book value"] / us_h["market value"])

# Calculating the short financial leverage
us_h["short leverage"] = us_h["shortdebt"] / us_h["totalassets"]

# Calculating f1 stock return
us_h_cc_1 = us_h.groupby([us_h.firm], as_index = False)[["annual stock return"]].shift(-1)
us_h = pd.merge(us_h, us_h_cc_1, left_index=True, right_index=True, how = "left")
us_h["annual stock return"] = us_h["annual stock return_x"]
us_h["f1 stock return"] = us_h["annual stock return_y"]

us_h = us_h.drop(columns = ["adjprice_x", "adjprice_y", "profit margin_x","profit margin_y", "market value_x","market value_y", "book value_x", "book value_y", "annual stock return_x", "annual stock return_y"])

us_h
```

#### ***Operating profit margin***

```python
# Selecting the column profit margin
us_h_service_industries_profit = us_h_service_industries[["profit margin", "q", "firm size", "firm"]]
us_h_service_industries_profit = us_h_service_industries_profit[~us_h_service_industries_profit.isin([np.nan, np.inf, -np.inf]).any(1)]
us_h_service_industries_profit
```

```python
# The median represets the profit margin's typical value. We can see in the description of the profit margin that the column presents outliers (min, max); thus, it is necessary to winsorize it.
us_h_service_industries_profit[["profit margin"]].describe()
```

- Design and run a multiple regression model to examine whether your selected factors and earnings per share deflated by price can explain/predict annual stock returns. You have to control for industry and firm size. To control for these variables you have to include them as extra independent variables in the model

#### ***Multiple regression model***

```python
# Dropping the rows which include nans, inf, -inf
variables_model = us_h_service_industries[["epsp", "medium firm", "big firm","profit margin", "book/market", "short leverage", "f1 stock return"]]
variables_model = variables_model[~variables_model.isin([np.nan, np.inf, -np.inf]).any(1)].reset_index()

us_h_service_industries = pd.merge(us_h_service_industries, variables_model, how = "inner")
```

```python
# Defining our independent and dependent variables
x = us_h_service_industries[["epsp", "medium firm", "big firm","profit margin", "book/market", "short leverage"]]
y = us_h_service_industries["f1 stock return"]
x
```

```python
# Verifying that the model has no multicollinearity in the independent variables.
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns

vif_data["VIF"] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]

# Since the Variance Inflation Factor is really small for all the independent variables, all these variables will be used for the model
print(vif_data)
```

```python
# Applying multiple regression model
X = sm.add_constant(x)
mkmodel = sm.OLS(y, X).fit()

print(mkmodel.summary())
```

- GENERAL RESULTS:
AFTER THE APPLICATION OF A MULTIPLE REGRESSION MODEL, WE CAN OBSERVE THAT THE R-SQUARED (THE ERROR) IS 2%, WHICH MEANS THAT THIS MODEL REPRESENTS A RELIABLE PREDICTION MODEL FOR THE ANNUAL STOCK RETURNS.

- INDEPENDENT VARIABLES ANALYSIS:
WE CAN OBSERVE THAT THERE ARE 3 POSITIVE COEFFICIENTS (CONSTANT, EPSP AND THE PROFIT MARGIN), WHICH MEANS THAT THEY HAVE A POSITIVE IMPACT OF MAGNITUDE X IN THE PREDICTION OF THE ANNUAL STOCK RETURN.
ON THE OTHER HAND, THERE ARE 4 NEGATIVE  COEFFICIENTS (MEDIUM FIRM, BIG FIRM, BOOK/MARKET AND SHORT LEVERAGE), WHICH MEANS THAT THEY HAVE A NEGATIVE IMPACT OF MAGNITUDE X IN THE PREDICTION OF THE ANNUAL STOCK RETURN.

*FOR THE VARIABLES: EPSP, MEDIUM FIRM, BIG FIRM, PROFIT MARGIN AND BOOK/MARKET*

WE CAN OBSERVE THAT THEIR T-VALUE IS GREATER THAN 2 OR LESS THAN -2. THE T-VALUE IS THE MEAN MINUS THE NULL HYPOTHESIS DIVIDED BY THE STANDARD ERROR, IN THIS CASE THE MEAN OF THE VARIABLES ARE THE COEFFICIENTS AND THE NULL HYPOTHESIS IS THE HYPOTHESIS WHERE THE COEFFICIENT OF THE VARIABLE IS ACTUALLY 0. THE RESULT OF THE T-VALUE TELLS US HOW MANY STANDARD DEVIATIONS IS OUR HYPOTHESIS AWAY FROM THE NULL HYPOTHESIS; LETS REMEMBER THAT FOR A NORMAL DISTRIBUTION, 2 STANDARD DEVIATIONS REPRESENT A 95% PROBABILITY UNDER THE CURVE, THUS, FOR A HYPOTHESIS TO BE APPROVED, THE T-VALUE MUST BE AT LEAST 2 (-2 FOR A NEGATIVE RELATION). 
THE P VALUE IS THE PROBABILITY OF THE NULL HYPOTHESIS TO BE CORRECT. AS WE CAN SEE, THE P-VALUE FOR THESE VARIABLES IS BASICALLY 0, WHICH MEANS THAT THE PROBABILITY OF THE COEFFCIENTS TO BE 0 IS BASICALLY 0.


*FOR THE VARIABLE: SHORT LEVERAGE*

THE T-VALUE IS GREATER THAN -2 AND LESS THAN 2, WHICH MEANS THAT THERE IS LESS THAN A 95% OF PROBABILITY THAT THE COEFFICIENT IS ACTUALLY THAT VALUE; THE P-VALUE OF 0.088 REPRESENTS THAT THERE IS AN 8.8% OF PROBABILITY THAT THE NULL HYPOTHESIS IS TRUE. SINCE THE T-VALUE OF THE COEFFICIENT OF THE SHORT LEVERAGE IS NOT LESS THAN 2, WE WILL DROP THE VARIABLE TO ADJUST THE MODEL.

```python
# Removing the short leverage independent variable
x = us_h_service_industries[["epsp", "medium firm", "big firm","profit margin", "book/market"]]

# Applying multiple regression model
X = sm.add_constant(x)
mkmodel = sm.OLS(y, X).fit()

print(mkmodel.summary())
```

WE CAN OBSERVE THAT THE R-SQUARED REMAIN WITH THE SAME VALUE (0.02) AND ALL THE OTHER COEFFICIENTS OF THE INDEPENDENT VARIBLES REMAIN BASICALLY THE SAME.

THE PREDICTED COEFFICIENTS REPRESENT THE MEAN CHANGE OF A DEPENDENT VARIABLE WHEN THE INDEPENDENT MOVES IN 1 UNIT. FOR EXAMPLE:
- FOR A 1 UNIT CHANGE IN THE BOOK MARKET RATE, THE STOCK RETURN FOR NEXT QUARTER IS GOING TO BE REDUCED BY \$0.1531
- FOR A 1 UNIT CHANGE IN THE PROFIT MARGIN, THE STOCK RETURN FOR NEXT QUARTER IS GOING TO BE INCREASED BY \$0.0449
